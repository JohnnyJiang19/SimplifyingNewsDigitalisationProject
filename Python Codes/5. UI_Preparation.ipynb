```python
## Import library
import pandas as pd

## Read data
df = pd.read_excel("1a. Final_Topic_Modelling_22_Topics_v2.xlsx",sheetname="Doc-Topic_Distribution")
article = pd.read_csv("Raw Data/Rawdata_articles.csv")

## Data transformation
arr_df = []
for i in range(2,24):
    df1 = df[["ID", df.columns[i]]]
    df1["Topic"]= df1.columns[1]
    df1.dropna(inplace=True)
    df2 = df1[["Topic","ID"]]
    arr_df.append(df2)
result = pd.concat(arr_df)
result.reset_index(drop=True, inplace=True)

## Merge two dataframe
df3 = pd.merge(result,article,how="left",left_on="ID",right_on="id")

## Extract useful columns
df4 = df3[["Topic","ID","title","publication"]]

## Read and clean summarization data
summary = pd.read_csv("4a. TFIDF_Summary.csv")
summary_split = summary["NO."].str.split(".", n=-1, expand=True)
summary["id"] = summary_split[0]
summary["id1"] = summary["id"].astype(int)

## Merge df4 and summarization dataframe
df5 = pd.merge(df4,summary,how="left",left_on="ID",right_on="id1")

## Extract useful columns
df6 = df5[["Topic","ID","title","publication","summary"]]

## Write to csv
df6.to_csv("5. UI_data_output.csv")
df6
```
